```python
import csv
csv_column=['Name','Address','age']
dicts=[{'Name': 'sita', 'Address': 'ktm', 'age': '89'},
       {'Name': 'rita', 'Address': 'thailand', 'age': '89'},
       {'Name': 'tiyup', 'Address': 'uk', 'age': '56'}]

f=open("data.csv",'w')
writer=csv.DictWriter(f,fieldnames=csv_column)
writer.writeheader()
for data in  dicts:
    writer.writerow(data)
```
